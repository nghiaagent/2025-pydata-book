pandas support dataframes and dataframes operations. Dataframes can be heterogeneous, unlike arrays.

```python
# Import packages
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

## Introduction to pandas Data Structures

There are 2 main structures, `Series` and `DataFrame`.

### Series
A series is a 1D-array-like object containing a single sequence of values (same type) and associated indices. They can be created from lists, arrays, or dictionaries.

```python
# Create via dictionary
obj = pd.Series({0: 4, 1: 7, 2: -5, 3: 3})

# Convert back to dictionary
obj.to_dict()
```

```python
# Create via list
obj = pd.Series([4, 7, -5, 3])
obj
```

Access the array and indices using associated methods.

```python
obj.index
```

```python
obj.array
```

The index can be set at the point of array creation. or be set afterwards

```python
# Modify previous series
obj.index = ["d", "b", "a", "c"]
obj
```

```python
# Create new series
obj2 = pd.Series([4, 7, -5, 3], index=["d", "b", "a", "c"])
obj2
```

Subset using either position of variable or by name.

```python
print(obj.iloc[[1, 2]])
print(obj.loc[["b", "a"]])
```

Boolean operations can be performed based on indices.

```python
"b" in obj
```

If both a dictionary and an index is provided at creation of series, the custom index will be created first, then valid dictionary data will be appended to the series.

```python
obj2 = pd.Series({"Ohio": 35000, "Texas": 71000, "Oregon": 16000, "Utah": 5000})

obj3 = pd.Series(
    {"Ohio": 35000, "Texas": 71000, "Oregon": 16000, "Utah": 5000},
    index=["California", "Ohio", "Oregon", "Texas"],
)

# Note that the order of values doesn't match with dictionary order, but with indices.
obj2
```

Arithmetic operations on series automatically aligns series by their index. Afterwards, the indices are also joined.

```python
obj4 = obj2 + obj3
obj4
```

Both the series data and its index can have names.

```python
obj4.name = "Population"
obj4.index.name = "State"
obj4
```

### DataFrame

A dataframe is an ordered, named collection of columns. It can be constructed via a dictionary of equal-length 1D arrays.

```python
data = {
    "state": ["Ohio", "Ohio", "Ohio", "Nevada", "Nevada", "Nevada"],
    "year": [2000, 2001, 2002, 2001, 2002, 2003],
    "pop": [1.5, 1.7, 3.6, 2.4, 2.9, 3.2],
}

frame = pd.DataFrame(data)
frame
```

Similar to series, a list of column names can also be passed to order the data.

```python
frame_ordered = pd.DataFrame(data, columns=["year", "state", "pop", "debt"])
frame_ordered
```

Each column can be selected via subsetting or as an attribute.

```python
frame_ordered[["year"]]
```

```python
frame_ordered.debt
```

Note that attribute-based subsetting only works if the column name does not conflict with methods and is a valid name.

```python
# This will call the method, not column 2
frame_ordered.pop
```

Retrieve rows via row indices using `loc` or `iloc` as with Series

```python
frame_ordered.loc[2]
```

Modify columns by assignment. If a series is assigned, the indices will be matched between the dataframe and series.

```python
frame_ordered["debt"] = np.random.standard_normal(6)
frame_ordered
```

```python
frame_ordered["debt"] = pd.Series([3, 5, 1], index=[2, 4, 5])
frame_ordered
```

The index and columns of DataFrames can have names.

```python
frame_ordered.index.name = "no"
frame_ordered.columns.name = "Data"
frame_ordered
```

DataFrames can be turned into arrays.

```python
frame_ordered.to_numpy()
```

## Essential Functionality

### Reindexing

Use reindexing to create a new dataframe / series from the current dataframe, with new indices following the usual syntax. 

```python
# Create Series
obj3 = pd.Series(["blue", "purple", "yellow"], index=[0, 2, 4])

# Expand indices
obj4 = obj3.reindex(np.arange(6))

# Expand indices, interpolate NAs
obj5 = obj3.reindex(np.arange(6), method="ffill")
print("obj3\n", obj3)
print("obj4\n", obj4)
print("obj5\n", obj5)
```

With dataframes, reindex rows with the `index` arg, reindex columns with the `columns` arg.

```python
# Example df
frame_ordered
```

```python
# Reindex
frame_ordered.reindex(index=[0, 1, 3, 4], columns=["year", "pop"])
```

### Dropping Entries from an Axis

Similar logic as reindex - pass index and columns labels to drop from data. Note that this does not mutate the original objects.

```python
# Drop on series
obj5.drop([0, 2, 4])
```

```python
# Drop on DataFrame
frame_ordered.drop(index=[2, 5], columns=["debt"])
```

### Indexing, Selection, and Filtering

Use the `loc` and `iloc` methods to subset Series and DataFrames based on indices.

```python
frame_ordered.iloc[1:5, 1:4]
```

```python
frame_ordered.loc[[0, 1, 2], ["state", "debt"]]
```

### Arithmetic and Data Alignment

Pandas automatically align Series and DataFrames by their indexes during math operations. However, during these operations, missing values propagate.

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=["a", "c", "d", "e"])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=["a", "c", "e", "f", "g"])
```

```python
s1 + s2
```

By performing additions using appropriate methods and the `fill` argument, missing values can be filled automatically (prior operation).

```python
s1.add(s2, fill_value=0)
```

#### Operations between DataFrame and Series

When a math operation is performed on a DataFrame on a Series, it is by default performed row by row. Pass the `axis` arg to methods to perform it by column instead.

```python
# Set up data
frame = pd.DataFrame(
    np.arange(12.0).reshape((4, 3)),
    columns=["b", "d", "e"],
    index=["Utah", "Ohio", "Texas", "Oregon"],
)

rows = pd.Series([4, 3, 2, 1], index=["Utah", "Ohio", "Texas", "Oregon"])

columns = pd.Series([3, 2, 1], index=["b", "d", "e"])

print(frame)
print(columns)
print(rows)
```

```python
# Perform operation per row
frame * columns
```

```python
# Perform per column
frame.add(rows, axis="index")
```

### Function Application and Mapping

NumPy ufuncs can be performed on each cell in a pandas object.

```python
# Set up data
df1 = pd.DataFrame(
    np.random.standard_normal(12).reshape((4, 3)),
    columns=["b", "d", "e"],
    index=["Utah", "Ohio", "Texas", "Oregon"],
)

df1
```

```python
# Set up data (Series)
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=["a", "c", "d", "e"])

s1
```

```python
# Perform ufunc on each element
np.abs(df1)
```

Use `apply` to perform a function on a column-wise or row-wise basis, treating each column / row as a 1D array.

```python
# Define function
def fun(x):
    return x.max() - x.min()
```

```python
# Perform across each row
df1.apply(fun, axis="rows")
```

```python
# Perform across each column
df1.apply(fun, axis="columns")
```

Many methods in dataframes are already applied column-wise / row-wise.

```python
df1.mean()
```

Element-wise functions performing like unary ufuncs can also be defined. Apply them on dataframes with `map`

```python
# Define function
def fun2(x):
    return f"{x:.2f}"


# Apply function
df1.map(fun2)
```

### Sorting and Ranking

#### Sorting

To sort by row or column index, use the `sort_index` method

```python
# Sort a series
s1.sort_index()
```

```python
# Sort a df by rows
df1.sort_index(axis="rows")
```

```python
# Sort a df by columns
df1.sort_index(axis="columns")
```

To sort by column values, use the `sort_values` method and pass in relevant columns (dataframes)

```python
df1.sort_values(["b", "e"])
```

```python
s1.sort_values()
```

#### Ranking
The `rank` method returns the indices of a Series or DataFrame such that the Series can be reordered by ascending order following the ranks.

```python
# Get rank
s1.rank()
```

```python
# Reorder by ranks
# Take 1 from ranks to get valid indices
s1.iloc[s1.rank() - 1]
```

For dataframes, reordering can occur over rows or columns.

```python
# Over rows (within columns)
df1.rank(axis="rows")
```

```python
# Over columns (within rows)
df1.rank(axis="columns")
```

### Axis Indexes with Duplicate Labels

Non-unique indexes makes life more difficult. Subsetting a series of dataframe with non-unique indices can produce unexpected datatypes.

```python
# Create data
## df1 unique
df1 = pd.DataFrame(
    np.random.standard_normal(12).reshape((4, 3)),
    columns=["b", "d", "e"],
    index=["Utah", "Ohio", "Texas", "Oregon"],
)

## df2 nonunique
df2 = pd.DataFrame(
    np.random.standard_normal(12).reshape((4, 3)),
    columns=["b", "d", "e"],
    index=["Utah", "Utah", "Texas", "Oregon"],
)

## s1 unique
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=["a", "c", "c", "e"])

## s2 nonunique
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=["a", "c", "d", "e"])
```

Subsetting a unique index in a series produces a single value, while a non-unique index produces a series.

```python
# Unique
s1["c"]
```

```python
# Non-unique
s2["c"]
```

Subsetting a unique index in a series produces a series, while a non-unique index produces a dataframe.

```python
# Unique
df1.loc["Utah"]
```

```python
# Non-unique
df2.loc["Utah"]
```
