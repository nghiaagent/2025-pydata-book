## Data Structures and Sequences

Common Python sequence types are tuples, lists, and dictionaries.

### Tuple

A tuple is a fixed-length, immutable list of objects.

````python
tup = (4, 5, 6)
tup
````

A tuple itself is immutable, but it can contain mutable objects (e.g. lists)

````python
# Create tuple containing a list at index 2
a = (4, 0, [1, 2])
a
````

````python
# Mutate the list at index 2 inside the tuple
a[2].append(3)
a
````

#### Unpacking tuples

To unpack a tuple, assign a similarly structured tuple of variable names to it.

````python
tuple = (4, 5, (6, 7))

(a, b, (c, d)) = tuple

print(a)
print(b)
print(c)
````

Note that the outer-most brackets are optional; meaning the above expression can look like below:

````python
tuple = (4, 5, 6)

a, b, c = tuple
````

### List

Lists are variable in length and mutable. They can be coerced from a tuple.

````python
# Create a list
a = [4, 5, 6]
a
````

````python
# Coerce a tuple into a list
tuple = (4, 5, 6)
a = list(tuple)
a
````

#### Adding and removing elements

Add an element to the end of the list using the `append()` method.

````python
a.append(7)
a
````

Insert an element at a desired index using the `insert()` method.

````python
a.insert(2, 5.5)
a
````

Remove an element at a desired index using the `pop()` method

````python
a.pop(2)
````

````python
a
````

Search and remove elements by value with the `remove` method.

````python
a.remove(7)
a
````

#### Sorting
Sort elements with the `sort` method :)

````python
a = [6, 7, 2, 5]
a.sort()
a
````

#### Slicing
Subset using the `[1st-index:last-index+1]` formula. Omit either the start or end index if they are the first or final index.

````python
a = [7, 2, 3, 7, 5, 6, 0, 1]
# Gets the 3rd element to the 5th element in list (index 4, one index before index 5)
a[2:5]
````

````python
# Gets the 3rd element to the final element in list
a[2:]
````

We can also subset using steps.

````python
# Gets every other element (0, 2, 4, 6, etc.)
a[::2]
````

````python
# Gets every element, starting from the end, going backwards
a[::-1]
````

### Dictionary

A dictionary is a collection of key-value pairs. It is a collection of ordered pairs (2-tuples, first element being key and second being value).

````python
dictionary = {"a": "first value", "b": [1, 2, 3, 4], 7: "an integer"}
````

Dictionaries are accessed via keys, not via indices

````python
# Access via key
dictionary["a"]
````

Add/modify values by accessing the respective keys.

````python
dictionary["a"] = "better first value"
dictionary[6] = "small integer"
dictionary
````

Remove values by deleting the respective key or with the `pop()` method

````python
dictionary.pop(6)
del dictionary["a"]
dictionary
````

Get the `keys` and `values` of a dictionary using the respective methods.

````python
dictionary.keys()
````

````python
dictionary.values()
````

### List, Set, Dictionary Comprehensions

#### List Comprehensions

It's like a map and a filter all in one.

This comprehension:
`[expr for value in collection if condition]`

is equivalent to the loop:
```
result = []
for value in collection:
    if condition:
        result.append(expr)
```

And the following in R:
```{r}
result <- collection %>%
    dplyr::filter(condition) %>%
    purrr::map(\(value) expr(value))
```

Either the condition or expression can be omitted, yielding just a `filter`-like or a `map`-like comprehension.

````python
# Create list
strings = ["a", "as", "bat", "car", "dove", "python"]
````

````python
# Turn all uppercase, but no filter, similar to a map
[x.upper() for x in strings]
````

````python
# Filter for strings longer than 2, similar to a filter
[x for x in strings if len(x) > 2]
````

````python
# Both filter for strings longer than 2 and map
[x.upper() for x in strings if len(x) > 2]
````

#### Dictionary Comprehensions

They are similar to list comprehensions but work on the unit of key-value pairs instead.


````python
dictionary = {"a": "first value", "b": "second value", 7: "an integer"}
dictionary
````

````python
# Process the above dictionary through a comprehension that does nothing
{key: value for (key, value) in dictionary.items()}
````

````python
# Turn all values uppercase like a map
{key: value.upper() for (key, value) in dictionary.items()}
````

````python
# Turn all values uppercase and filter based on key
{
    key: value.upper()
    for (key, value) in dictionary.items()
    if (isinstance(key, str) == True)
}
````

#### Set Comprehensions

They are also similar to list comprehensions, using `{}` brackets instead of `[]`. Set comprehensions can create sets from input lists.

````python
set_example = [1, 2, 1, 2, 3, 4, 2, 6]
print(set_example)
````

````python
# Create set then convert all values to strings
{str(x) for x in set_example}
````

#### Nested list comprehensions

Say we want to apply maps and filters to each value in a list of lists - this is how. Be careful of the output structure you want.

````python
# Create nested list
list_nested = [
    ["John", "Emily", "Michael", "Mary", "Steven"],
    ["Maria", "Juan", "Javier", "Natalia", "Pilar"],
]
````

````python
# Comprehension to filter for names starting with M producing a flattened list:
[x.upper() for list in list_nested for x in list if x[0] == "M"]
````

````python
# Comprehension  producing a nested list:
[[x.upper() for x in list if x[0] == "M"] for list in list_nested]
````

## Functions

### Namespace

A function can access and mutate variables outside of its namespace, but any variables created inside its namespace are not kept.

````python
# Example: output is created inside function. output is not returned.
def fun():
    output = []
    for i in range(5):
        output.append(i)

fun()

# Check if output exists
try:
    output
except NameError:
    print("not defined")
````

````python
# Example: output is created outside function and is mutated by function
def fun():
    for i in range(5):
        output.append(i)

output = []
fun()

# Check if output exists
try:
    output
except NameError:
    print("not defined")
else:
    print(output)

# Subsequent function calls continue to mutate output
fun()

print(output)
````

### Returning multiple values

A function can return multiple values (via tuples).

````python
def fun(x):
    x = x
    y = x + 1
    z = y + 1
    return (x, y, z)

(a, b, c) = fun(1)
print(a, b, c)
````

### Generators

Generators doesn't output direct results, but an object that will produce results when requested.

````python
# A function that generate squares straight away
def squares(n):
    out = []
    for i in range(1, n + 1):
        out.append(i**2)
    return out

squares(5)
````

````python
# A generator function that generate squares when requested
def squares(n):
    for i in range(1, n + 1):
        yield i**2

# Use function to create a generator
out = squares(5)
print(out)

# Print output of generator
print(list(out))
````

#### Generator expressions
Similar to list expressions, there are also generator expressions.

````python
# Generator function form
def squares(n):
    for i in range(1, n + 1):
        yield i**2

out_1 = squares(10)

# Generator expression form
out_2 = (x ** 2 for x in range(1, 11))

# Prove that output are the same
list(out_1) == list(out_2)
````

## Files and the OS

To open a file, use `open()`. Supply an encoding type as best practice.

````python
file = open("../examples/segismundo.txt", encoding = "utf-8")
````

The file is treated like a list of lines and can be iterated line-wise.

````python
for line in file:
    print(line)
````

Once we are done with the file, close it.

````python
file.close()
````

All the interactions with a file can be wrapped inside a `with` statement.

````python
# Extract lines imperatively:
file = open("../examples/segismundo.txt", encoding="utf-8")
lines_1 = [x.rstrip() for x in file]
file.close()

# Using a with statement
with open("../examples/segismundo.txt", encoding="utf-8") as file:
    lines_2 = [x.rstrip() for x in file]

# Output are the same
lines_1 == lines_2
````
