Similar to R functions, when objects are passed into a function, a new copy is made.
Unlike R functions, the initial (outer) object is automatically modified.

```python
# Define a function to append data to list
def append_element(list, element):
    list.append(element)


# Construct example data
data = [1, 2, 3]

# Append 4 to data
append_element(data, 4)

# Show data has been modified even though there has been no explicit assignment.
data
```

#### Dynamic references, strong types

Similar to R, variables in Python do not themselves contain data, they are references to objects. Type coercion / casting is very limited.

```python
# a is bound to an integer
a = 5
type(a)
```

```python
# a is bound to a string
a = "Hello"
type(a)
```

#### Duck typing
Due to the OOP nature of Python, it's possible to check if an object's attribute is "close enough" by checking its methods. For example, we can check if an object is iterable by checking its `iter()` method.

```python
# Define function to check
def isiterable(x):
    try:
        iter(x)
        return True
    except TypeError:
        return False
```

```python
# Check
isiterable("Hello")
```

```python
# Note: In Python, a string is a list of characters. That's why they are iterable.
"Hello"[0:2]
```

```python
# Check if a list is iterable
isiterable([1, 2, 3])
```

```python
# Check if a number is iterable
isiterable(5)
```

#### Binary operators and comparisons

Use `is` and `is not` to check if two variables reference the same object.

Note that `list()` creates an identical copy of an object bound to the desired variable.

```python
a = [1, 2, 3]
# Refer to the original list with a new variable
b = a

# Create a copy
c = list(a)
```

```python
# Show that b is same as a
a is b
```

```python
# Show that c is not the same object as a
a is c
```

```python
# Show that c has the same content as a (but is a different object)
a == c
```

#### Mutable and immutable objects
Most objects (list, arrays, etc.) in Python are mutable.

Some are immutable, such as strings and tuples (similar to lists).

```python
# Create a list
a = ["foo", 2, [4, 5]]
a
```

```python
# Mutate it
a[2] = (3, 4)
a
```

```python
# Create tuple (identical content as above)
b = ("foo", 2, [4, 5])
b
```

```python
# Mutate it (not!)
b[2] = (3, 4)
```
